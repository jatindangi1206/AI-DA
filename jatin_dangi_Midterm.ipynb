````python
# Quesion 1 - Part (a)



import numpy as np
import matplotlib.pyplot as plt

# Number of points
n = 1000

# Generate random x and y coordinates
x = np.random.rand(n)
y = np.random.rand(n)

# Choose k and p
k = 150
p = k / 50

# Calculate prices based on euclidean distance from origin
prices = k * np.sqrt(x**2 + y**2)

# Ensure minimum price constraint
prices = np.where(prices < p, p, prices)

# Plot
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c=prices, cmap='Set1')
plt.colorbar(label='Price')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('House Prices based on Euclidean Distance from Origin')
plt.show()

k, p
````

````python
# Quesion 1 - Part (b)


# Calculate standard deviation of the signal (prices)
std_signal = np.std(prices)

# Choose an SNR value between 2 and 3
SNR = 2.5

# Calculate standard deviation of the noise
std_noise = std_signal / SNR

# Generate the noise
noise = np.random.normal(0, std_noise, n)

# Add noise to the prices
noisy_prices = prices + noise

# Ensure that the minimum price is positive and adjust if necessary
noisy_prices = np.where(noisy_prices < p, p, noisy_prices)

# Ensure that all prices are positive
noisy_prices = np.maximum(noisy_prices, 0.1)

# Plot histogram of the error
plt.figure(figsize=(10, 6))
plt.hist(noise, bins=50, color='blue', alpha=0.7)
plt.title('Histogram of the Noise (Error)')
plt.xlabel('Error Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Scatter plot for noisy prices
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c=noisy_prices, cmap='Set1')
plt.colorbar(label='Price with Noise')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('House Prices with Noise based on Euclidean Distance from Origin')
plt.show()

min_noisy_price = np.min(noisy_prices)
min_noisy_price
````

````python
# Question 1 - Part (c)

# Calculate the Euclidean distance for each house
F1 = np.sqrt(x**2 + y**2)
F2 = np.zeros_like(F1)

# Store dataset
data_b = np.c_[noisy_prices.T, F1.T, F2.T]

# Plot
plt.figure(figsize=(10, 8))
plt.scatter(F1, noisy_prices, color='purple', alpha=0.6)
plt.xlabel('Euclidean Distance from Origin (F1)')
plt.ylabel('Price with Noise (P)')
plt.title('House Prices with Noise vs. Euclidean Distance from Origin')
plt.grid(True)
plt.show()

data_b.shape
````

````python
# Question 1 - Part(d)


# Construct the X matrix with an added column of ones for the intercept
X = np.c_[np.ones(n), F1, F2]

# Y is the vector of noisy prices
Y = noisy_prices

# Calculate the pseudo inverse of X
pseudo_inv_X = np.linalg.pinv(X)

# Compute the coefficients
beta = np.dot(pseudo_inv_X, Y)

# Extract the estimated intercept and slope
intercept, slope, _ = beta

intercept, slope
````

````python
# Question 1 - Part (e)


# Calculate correlation between P and F1
corr_P_F1 = np.corrcoef(noisy_prices, F1)[0, 1]

# Calculate correlation between P and F2 (This will be essentially 0 since F2 is a constant vector of zeros)
corr_P_F2 = np.corrcoef(noisy_prices, F2)[0, 1]

# Plotting
plt.figure(figsize=(10, 8))
plt.scatter(F1, noisy_prices, color='purple', alpha=0.6, label='Data Points')
plt.plot(F1, intercept + slope * F1, color='red', label='Fitted Line')
plt.xlabel('Euclidean Distance from Origin (F1)')
plt.ylabel('Price with Noise (P)')
plt.title('House Prices with Noise vs. Euclidean Distance from Origin with Fitted Line')
plt.legend()
plt.grid(True)
plt.show()

corr_P_F1, corr_P_F2
````

````python
# Question 1 - Part (f)

# Calculate the direction vector for the landfill site
landfill_vector = np.array([1/2, 1/2])
landfill_magnitude = np.linalg.norm(landfill_vector)

# Calculate angles for each point with respect to the origin and the landfill direction
angles = np.arccos(np.dot(np.c_[x, y], landfill_vector) / (np.linalg.norm(np.c_[x, y], axis=1) * landfill_magnitude))
angles_deg = np.degrees(angles)

# Define the cone of +/- 15 degrees around the direction of landfill site
cone_min = np.degrees(np.arccos(np.dot([1, 0], landfill_vector) / landfill_magnitude)) - 15
cone_max = np.degrees(np.arccos(np.dot([1, 0], landfill_vector) / landfill_magnitude)) + 15

# Determine prices based on whether point is inside or outside the cone
prices_in_cone = (angles_deg >= cone_min) & (angles_deg <= cone_max)
adjusted_prices = np.where(prices_in_cone, 0.5 * k * np.sqrt(x**2 + y**2), k * np.sqrt(x**2 + y**2))

# Add noise to the prices
noise = np.random.normal(0, std_noise, n)
noisy_adjusted_prices = adjusted_prices + noise

# Ensure that the minimum price is positive and adjust if necessary
noisy_adjusted_prices = np.where(noisy_adjusted_prices < p, p, noisy_adjusted_prices)
noisy_adjusted_prices = np.maximum(noisy_adjusted_prices, 0.1)

# Store dataset
data_f = np.c_[noisy_adjusted_prices.T, F1.T, F2.T]

# Scatter plot for noisy prices with adjusted model
plt.figure(figsize=(10, 8))
plt.scatter(x, y, c=noisy_adjusted_prices, cmap='Set1')
plt.colorbar(label='Adjusted Price with Noise')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Adjusted House Prices with Noise based on Euclidean Distance and Landfill Site Influence')
plt.show()

min_noisy_adjusted_price = np.min(noisy_adjusted_prices)
min_noisy_adjusted_price

````

````python
# Question 2 - Part (a)

def sigmoid(z):
    """Compute the sigmoid function."""
    return 1 / (1 + np.exp(-z))

def compute_cost(y_hat, y):
    """Compute the mean squared error."""
    m = len(y)
    return 0.5 * np.sum((y_hat - y)**2) / m

def stochastic_gradient_descent(X, y, alpha, num_iterations):
    """
    Implement Stochastic Gradient Descent for logistic regression.

    Parameters:
    - X: Input features
    - y: Target variable
    - alpha: Learning rate
    - num_iterations: Number of iterations

    Returns:
    - theta: Parameters after optimization
    - cost_history: Cost function's values during iterations
    """

    # Initialize parameters
    theta = np.zeros(2)

    m = len(y)
    cost_history = []

    for _ in range(num_iterations):
        # Randomly select an instance
        i = np.random.randint(0, m)

        # Compute the prediction for the chosen instance
        z = theta[0] + theta[1] * X[i]
        y_hat = sigmoid(z)

        # Compute the gradients for theta_0 and theta_1
        gradient = np.array([(y_hat - y[i]), (y_hat - y[i]) * X[i]])

        # Update the parameters
        theta -= alpha * gradient

        # Compute the cost for the updated parameters and store in history
        y_hat_all = sigmoid(theta[0] + theta[1] * X)
        cost = compute_cost(y_hat_all, y)
        cost_history.append(cost)

    return theta, cost_history

# Example usage with dummy data
X_dummy = np.array([0.5, 1.5, 2.5, 3.5, 4.5])
y_dummy = np.array([0, 1, 0, 1, 0])

alpha = 0.01
num_iterations = 1000

theta_optimized, cost_history = stochastic_gradient_descent(X_dummy, y_dummy, alpha, num_iterations)
theta_optimized, cost_history[-5:]  # Displaying the last 5 cost values for brevity
````

````python
# Question 2 - Part (b)

# Generate X values
np.random.seed(42)
X_data = np.random.rand(100) * 10

# Initialize theta values such that z becomes large negative for the given X
theta_initial = [-50, -5]

# Compute y values
z_data = theta_initial[0] + theta_initial[1] * X_data
y_data = sigmoid(z_data)

# Introduce some noise to y
noise = np.random.binomial(1, 0.05, len(y_data))  # Some random noise
y_data_noisy = np.where(noise, 1 - y_data, y_data)

# Run SGD and print theta updates
def print_sgd_updates(X, y, alpha, num_iterations, theta_initial):
    """
    A variant of the SGD function that prints theta updates.
    """
    theta = np.array(theta_initial)
    m = len(y)
    theta_history = [theta_initial]

    for _ in range(num_iterations):
        i = np.random.randint(0, m)
        z = theta[0] + theta[1] * X[i]
        y_hat = sigmoid(z)

        gradient = np.array([(y_hat - y[i]), (y_hat - y[i]) * X[i]])
        theta -= alpha * gradient

        theta_history.append(list(theta))

    return theta_history

theta_history = print_sgd_updates(X_data, y_data_noisy, 0.01, 10, theta_initial)
theta_history
````
